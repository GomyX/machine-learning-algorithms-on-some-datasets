# **Lists**


Lists are the basic containers for grouping data together in Python. <br>
Lists are **mutable**.

```python
foo = [1,4,9,16]
print(foo)

foo[0] = 'Python' #elements of the list do not necessarily have the same type
print(foo)

foo[0:2] = ['PHP', 'C++'] #assign multiple items
print(foo)

bar = [[1,2],'Python',1]
print(bar)
```

Lists support indexing and slicing <br>
 List:&emsp;1&ensp;4&ensp;3&ensp;9 <br>
 Index: 0&ensp;1&ensp;2&ensp;3 <br>
 &emsp; &emsp;&ensp;-4 -3 -2 -1

```python
foo = [1,4,3,9]
print(foo) 

print(foo[2:],foo[-4:-2]) # supports indexing and slicing

print(foo[2:]+foo[-4:-2])

print(foo[::-1]) # read the list in a reversed order
```

Lists also support methods like strings. Find more [here.](https://www.w3schools.com/python/python_lists_methods.asp)

```python
example = [4,2,5,1]

example.sort() # sorts the list 
print(example)
```

Adding items to a list

```python
nums = [4,2,5,1]

nums.append(50) # appends an element to the end of the list
print(nums)

nums.insert(2, 100) #insert element at a certain index
print(nums)

fruits = ['Apple', 'Banana', 'Watermelon'] 
nums.extend(fruits) #add elements in an array to another
print(nums)

nums.remove("Banana") #remove element
print(nums)

nums.clear()
print(nums)
```

```python
#using lists as stacks - search how to use lists as queues
foo = [1,2,3]
print(foo.pop(),foo) #pop last element

print(foo.pop(1),foo) #pop a certain element
```

List-to-list assignment is a shallow copy => you create a reference to the list

```python
foo = [1,4,9,16]
example = foo #this is a shallow copy!
# Changing an element in 'example' changes the same element in 'foo'
example[0] = 'Not Python'
print(example,foo)
```

Python has plenty of built-in libraries besides the standard library. <br>
You import a library using the `import libraryname` statement

```python
import copy as cp
foo = [1,4,9,16]
example = cp.deepcopy(foo)
example[0] = 'new copy'
print(example,foo)
# You can also copy using a loop by creating an empty list and appending the contents of the other list to it.
```

# **Tuples**

The main difference between tuples and lists is that tuples are **immutable**. <br> Besides immutability, why use tuples over lists? 
<br> 

*   Tuples are **faster** than lists as lists are created using 2 blocks of memory; a fixed one with all the object information and a variable sized block of memory for possible future allocations. On the other hand, tuples are created using a single fixed block of memory which makes creating a tuple faster than creating a list.
*   Indexing is faster in tuples as it follows fewer pointers due to the reason mentioned in the previous bullet point.
*   Tuples also make your code safer (write-protected).

If you are going to define a constant set of elements, use tuples instead of lists.

```python
foo = (1,2,3) #any number of elements
bar = (4,5,'six') #does not have to be of a certain type

print(foo)
print(bar)
print(foo+bar) #add tuples
```

```python
foo = (1,2,3) 
foo[0]
```

```python
foo[1] = 5 # You can't do that -> immutable
```

```python
#add elements to a tuple
example = foo
example += ('four','five') # + is the concatenation operator
print(example is foo)
print(foo)
print(example)
```

```python
#delete from a tuple
foo = (1,2,3) 
foo_list = list(foo)
foo_list.remove(1)
foo = tuple(foo_list)
print(foo)
```

# **Dictionaries** 
It is best to think of dictionaries as key-value pairs => age: 12

```python
#creating a dictionary
bar = {'Ahmed': 1074,'Aya':1075}
print(bar)

#Another way of creating a dictionary is using a list of tuples where each tuple represents a key-value pair
key_value_pairs = [('Ahmed',1074),('Aya', 1075)]
foo = dict(key_value_pairs)
print(foo)
```

```python
#Access dict values
bar = {'Ahmed': 1074,'Aya':1075}
print(bar['Ahmed'])
```

```python
#nested dicts
students = {
  "student_1" : {
    "name" : "John",
    "Grade" : 'A'
  },
  "student_2" : {
    "name" : "May",
    "Grade" : 'B+'
  },
  "student_3" : {
    "name" : "Ahmed",
    "year" : 'C'
  }
}

print(students)
print(students['student_3'])
```

**For Loops** <br>
Write a code to print the first 10 digits of the fibonacci sequence

```python
a=1
for i in range(10):
    a+=1
    print('i = {} ,, a= {}'.format(i,a))
```

```python
for i in range(0,10,2):
    print(i)
```

**While loops**

```python
i = 1
a = 0
b = 1
while i<10:
    i+=1
    a,b = a+i, b+i # multiple assignment
    print('i = {}, a= {}, b= {}'.format(i,a,b))
```

For loops are particularly powerful in Python as they can loop on any iterable 

```python
#loop on a list
foo = ['Intro','to','Python']
for word in foo:
  print(word)
```

```python
for i in range(len(foo)):
  word = foo[i]
  print(word)
```

```python
bar = [[4,5,6],[7,7,9],[1,2,3]]
for sublist in bar:
    for element in sublist:
        print(element)
```

Looping over a dictionary

```python
bar = {'Ahmed': 1074,'Aya':1075}
for k,v in bar.items():
  print(k,v)
```

```python
#loop on two lists at once
a=[1,2,3,4]
b=[1,2,3,4]
for i,j in zip(a,b):
    print(i+j)
```

**Break and continue statements are similar to ones in C**

```python
foo = ['Intro','to','Python']
for word in foo:
    if word == 'to':
        break
    print(word)
```

```python
foo = ['Intro','to','Python']
for word in foo:
    if word == 'to':
        continue
    print(word)
```

# **Functions**

```python
def fib(n=10):
    ''' A function that computes the nth number in a fibonacci sequence
        fibonacci seq = 0,1,1,2,3,5,8,...etc
    '''
    a = 0
    b = 1
    for i in range(n-1):
        a,b = b,a+b
    return a
    
print(fib(),'-',fib(5))
```
